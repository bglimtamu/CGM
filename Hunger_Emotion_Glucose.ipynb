```python
# 통계 계산
stats = {
    "Min": [df["Session Compliance %"].min()],
    "Med": [df["Session Compliance %"].median()],
    "Mean": [df["Session Compliance %"].mean()],
    "SD": [df["Session Compliance %"].std()],
    "Max": [df["Session Compliance %"].max()],
}

# 표 생성
stats_df = pd.DataFrame(stats)

# 결과 출력
print(stats_df)
```

```python
from nbconvert.exporters import HTMLExporter
from nbconvert.preprocessors import TagRemovePreprocessor
import nbformat

# 태그 기반 필터 설정
def export_notebook(input_notebook, output_html, tag_to_include):
    # Notebook 읽기
    with open(input_notebook) as f:
        notebook = nbformat.read(f, as_version=4)
    
    # 태그 제거 프로세서 설정
    preprocessor = TagRemovePreprocessor()
    preprocessor.remove_cell_tags = [tag_to_include]
    preprocessor.enabled = True
    
    # HTML Exporter 설정
    html_exporter = HTMLExporter()
    html_exporter.preprocessors = [preprocessor]
    
    # HTML로 변환
    body, _ = html_exporter.from_notebook_node(notebook)
    with open(output_html, "w") as f:
        f.write(body)
    print(f"HTML 파일이 생성되었습니다: {output_html}")

# 실행: 파일 이름과 태그를 입력하세요
export_notebook("input_notebook.ipynb", "output.html", tag_to_include="export")
```
