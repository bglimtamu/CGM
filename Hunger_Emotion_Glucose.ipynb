```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import os
import pandas as pd
# 엑셀 파일 불러오기
df = pd.read_excel('/Users/bg.lim/Downloads/Texas A&M_Agrilife/Projects/1.CGM/Pilot/Datas/EMA/BG_EMA/75_EMA_Comp.xlsx', engine='openpyxl')  # 파일 경로 지정

df
```

```python
# 통계 계산
stats = {
    "Min": [df["Session Compliance %"].min()],
    "Med": [df["Session Compliance %"].median()],
    "Mean": [df["Session Compliance %"].mean()],
    "SD": [df["Session Compliance %"].std()],
    "Max": [df["Session Compliance %"].max()],
}

# 표 생성
stats_df = pd.DataFrame(stats)

# 결과 출력
print(stats_df)
```

```python
from tabulate import tabulate

# 통계 데이터
stats = {
    "Statistic": ["Min", "Median", "Mean", "SD", "Max"],
    "Value": [
        df["Session Compliance %"].min(),
        df["Session Compliance %"].median(),
        df["Session Compliance %"].mean(),
        df["Session Compliance %"].std(),
        df["Session Compliance %"].max(),
    ],
}

# 표 생성
stats_df = pd.DataFrame(stats)

# tabulate로 표 출력
print(tabulate(stats_df, headers="keys", tablefmt="grid"))
```
