# Model 4: 30+60+90의 평균값을 하나의 변수로, Model 2(30min)가 제일 AIC, BIC 낮게나옴

```python
import pandas as pd

# 엑셀 파일 불러오기
df = pd.read_excel('/Users/bg.lim/Downloads/Texas A&M_Agrilife/Projects/1.CGM/Pilot/Datas/Preprocessing/3.within_timepoint/Timepoint_null.xlsx', engine='openpyxl')  # 파일 경로 지정
df
```

```python
import pandas as pd
import statsmodels.formula.api as smf
```

```python
print(df.columns)
```

```python
df['CGM_combined_mean'] = df[['CGM_30min_mean', 'CGM_60min_mean', 'CGM_90min_mean']].mean(axis=1)
```

```python
df
```

```python
# Step 6: 확인 (정리된 데이터)
print(df[['StudyID', 'Date', 'within_day_num', 'CGM_combined_mean']].head())
```

```python
# Step 7: 혼합효과 선형모델 실행 (CGM_combined_mean를 종속변수로 사용)
model = smf.mixedlm(
    "CGM_combined_mean ~ hungry + bored + How_angry + DailyAvgLoneliness + How_sad + How_tired",
    data=df,
    groups=df["within_day_num"]  # 같은 날 그룹화된 데이터를 랜덤 효과로 사용
)

# Step 8: 모델 적합 및 결과 출력
result = model.fit()
print(result.summary())

# Step 9: AIC와 BIC 출력
print(f"AIC: {result.aic}")
print(f"BIC: {result.bic}")
```
